El ***análisis de sentimientos*** es una tarea del procesamiento del lenguaje natural que permite identificar la *polaridad* de un texto, es decir, si es *positivo*, *negativo* o *neutro*. Esta tarea es importante porque permite **comprender** la **opinión** de los **usuarios**, detectar el *discurso* de *odio* y *mejorar* la *experiencia* del *usuario*.

***Pysentimiento*** es una biblioteca de análisis de sentimientos basada en transformers para el idioma español. Se basa en la implementación de BERT en español (BETO) y los datos de la versión 2020 del Taller de Análisis de Sentimientos (TASS) de la Sociedad Española de Procesamiento del Lenguaje Natural (SEPLN).

```python
print("Español:", accuracy_score(Y_validation_N, Y_ClasificacionBA_N))
print("Varios Idiomas:", accuracy_score(Y_validation_T, Y_ClasificacionBA_T))
```

```python
auc_list_N = []
auc_list_T = []
for i in range(n_classes_N):
    fpr_i, tpr_i, _ = roc_curve(y_test_bin_N[:, i], y_score_N[:, i])
    auc_i = auc(fpr_i, tpr_i)
    auc_list_N.append(auc_i)
for i in range(n_classes_T):
    fpr_i, tpr_i, _ = roc_curve(y_test_bin_T[:, i], y_score_T[:, i])
    auc_i = auc(fpr_i, tpr_i)
    auc_list_T.append(auc_i)
```

```python
conf_matrix = confusion_matrix(Y_validation_N, Y_ClasificacionBA_N)
accuracy_list_N = []
precision_list_N = []
sensitivity_list_N = []
specificity_list_N = []
f1_list_N = []

for i in range(conf_matrix.shape[0]):
    vp = conf_matrix[i, i]
    fn = np.sum(conf_matrix[i, :]) - vp
    fp = np.sum(conf_matrix[:, i]) - vp
    vn = np.sum(conf_matrix) - vp - fn - fp
    
    accuracy_i = (vp + vn) / (vp + fp + fn + vn)
    precision_i = vp / (vp + fp)
    sensitivity_i = recall_i = vp / (fn + vp)
    specificity_i = vn / (fp + vn)
    f1_i = (2 * precision_i * recall_i) / (precision_i + recall_i)
    
    accuracy_list_N.append(accuracy_i)
    precision_list_N.append(precision_i)
    sensitivity_list_N.append(sensitivity_i)
    specificity_list_N.append(specificity_i)
    f1_list_N.append(f1_i)

print("Accuracy por clase:", accuracy_list_N)
print("Precision por clase:", precision_list_N)
print("Sensitivity por clase:", sensitivity_list_N)
print("Specificity por clase:", specificity_list_N)
print("F1 Score por clase:", f1_list_N)
```
